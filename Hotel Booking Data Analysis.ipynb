The accompanying bar shows the percentage of reservation that are canceled and those that are not.It is obvious that there are still a significant number of reservation that have not been canceled.There aer 37% of clients who canceled there reservation,which has a signigicant impact on the hotel'S earning

```python
plt.figure(figsize=(8,4))
ax=sns.countplot(x="hotel",hue='is_canceled',data=df)

plt.title("Reservation Status in different hotel")
plt.xlabel("hotel")
plt.ylabel("Number of reservation")
plt.legend(["Not Cancelled","Chancelled"])
plt.show()
```

In comparison to resort hotels,city hotels have more booking.It's possible that resort hotel are more expensive than those in city

```python
resort_hotel=df[df["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Avarage Daily Rate of Resort and City Hotel",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel["adr"],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel["adr"],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
```

The line graph avobe shows that,on certain days, the average daily rate for a city hotel isless then that of a resort hotel.nd on the other days,it is even less. It gose without saying that weekends and holidays may see a rice in resort hotel rates.

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(18,8))
ax=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
plt.title=('Reservation Status Per Month')
plt.xlabel('month')
plt.ylabel('Number of Reservation')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

We have developed the grouped bar graph to analyze the months with the higest and lowest reservation levels according to reservation status.As can be seen, both the number of confirmed reservations and number of canceled reservation are largest in the month of August.Wheres January is the month with the most canceled reservations.

```python
plt.figure(figsize=(15,8))
plt.title=('ADR Per Month')
sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())

#plt.legend(fontsize=20)
plt.show()
```

This bar graph demonstrates that cancellations are most common when prices are greatest and are least common when they arelowest.Therfore,The cost of the accommodation is solely responsible for the cancellation.

```python
cancelled_data=df[df["is_canceled"]==1]
top_10_country=cancelled_data["country"].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title=("Top 10 Country with reservation canceled")
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.show()
```

In this pie chart show that which country has the higest reservation canceled. The top country is Portugal with the higest number of cancellations.

```python
cancelled_df_adr=cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

not_cancelled_data=df[df["is_canceled"]==0]
not_cancelled_df_adr=not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

plt.figure(figsize=(20,6))
#plt.title("Avarage Daily Rate ",fontsize=30)
plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="not_canceled")
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="canceled")
plt.legend()
plt.show()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr["reservation_status_date"]>"2016")&(cancelled_df_adr["reservation_status_date"]<"2017-09")]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr["reservation_status_date"]>"2016")&(not_cancelled_df_adr["reservation_status_date"]<"2017-09")]
plt.figure(figsize=(20,6))
#plt.title("Avarage Daily Rate ",fontsize=30)

plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="not_canceled")
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="canceled")
plt.legend()
plt.show()
```

As seen in thegraph, reservations are canceled when the average daily rate is higher than when it is not canceled.It clearly proves all the above analysis,that the higher price leades to higher cancellation.
